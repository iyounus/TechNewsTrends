```python
from __future__ import division
import numpy as np
import pandas as pd
import pickle as pkl
import os
import json

import datetime
import unicodedata

import collections

from wordcloud import WordCloud

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF

from nltk import WordNetLemmatizer, sent_tokenize, word_tokenize
from nltk.corpus import stopwords

from utils import tokenize

from datetime import date

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pkl.load( open("data/data_all.pkl"))
df.pub_date = pd.to_datetime(df.pub_date)
df['pub_year'] = df.pub_date.map(lambda x : date.isocalendar(x)[0])
df['pub_week'] = df.pub_date.map(lambda x : date.isocalendar(x)[1])
df['pub_month'] = df.pub_date.map(lambda x : x.month)

# 0 -> Monday of the pub_week
df['pub_week_date'] = df.pub_date.map(lambda x : x.date() + datetime.timedelta(0-x.date().weekday()))
df['pub_week_date_str'] = df.pub_date.map(lambda x : (x.date() + 
                                                      datetime.timedelta(0-x.date().weekday())).strftime("%Y-%m-%d"))

df.info()

vectorizer, vectors = pkl.load( open("data/vectorizer.pkl", "rb") )
nmf = pkl.load( open("data/model_45.pkl", "rb"))
```

```python
W = nmf.components_#[topicN]
A = vectors.dot(W.T)

df['topic'] = list(np.argmax(A, axis=1))
df['weight'] = list(np.max(A, axis=1))

df = df[df['weight']>0.5]

# now sort topics w.r.t number of articles per topic
# this is just renaming the topic
dg = df[['topic','headline']].groupby('topic')
x = sorted(dg.groups.keys())
y = [len(dg.groups[i]) for i in x]
m = list(np.argsort(y)[::-1])
d = { j : x[i] for i, j in enumerate(m)}

df['topic_sorted'] = df['topic'].map(lambda x : d[x])

df[['topic','topic_sorted','weight','pub_date','pub_week','pub_week_date','headline']].head()
#print d
```

```python
def PlotNarticles(col):
    dg = df[[col,'headline']].groupby(col)
    #x = sorted(dg.groups.keys())
    x = sorted(dg.groups.keys())
    y = [len(dg.groups[i]) for i in x]

    #print x
    #print y
    plt.figure(figsize=(12,4))
    plt.plot(x,y,'o-');
    
#PlotNarticles('topic')
PlotNarticles('topic_sorted')
PlotNarticles('pub_week_date')
```

```python
def articles_week_dict(): #articles per week
    dg = df[['pub_week_date_str','headline']].groupby('pub_week_date_str')
    return dg.size().to_dict()


def draw(topic):
    d = articles_week_dict()

    #topic = 1
    dftopic = df[['pub_week_date','headline']][df['topic_sorted']==topic]
    dg = dftopic.groupby('pub_week_date')
    x = sorted(dg.groups.keys())
    y = [len(dg.groups[i]) for i in x]

    #print x
    #print y
    #d = dict(zip(x,y))

    plt.figure(figsize=(15,5))
    plt.plot(x,y,'o-');

    
def draw2(topic):
    x = []
    y1 = []
    y2 = []
    col0 = ['topic_sorted','pub_week','pub_week_date','web_url']
    d = articles_week_dict()
    for pub_week_date in df['pub_week_date'].order().unique().tolist():
        pub_week_date_str = pub_week_date.strftime("%Y-%m-%d")

        cond = "topic_sorted == " + str(topic) + \
               " & pub_week_date_str == '" + pub_week_date_str + "'"

        dg = df.query(cond).sort(['weight'], ascending=[0])[col0]

        urls = dg['web_url'].values.tolist()

        y1.append(len(urls))
        y2.append(len(urls)/d[pub_week_date_str])
        x.append(pub_week_date)

    plt.figure(figsize=(15,8))
    plt.subplot(2,1,1)
    plt.plot(x,y1,'o-');
    plt.subplot(2,1,2)
    plt.plot(x,y2,'o-');
    return x, y1, y2
```
