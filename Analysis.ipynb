```python
import numpy as np
import pandas as pd
import pickle as pkl
import os
import json

import datetime
import unicodedata

from wordcloud import WordCloud

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF

from nltk import WordNetLemmatizer, sent_tokenize, word_tokenize
from nltk.corpus import stopwords

from topic_ana import featurize

from datetime import date

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pkl.load( open("data/nyt_data.pkl"))
df.pub_date = pd.to_datetime(df.pub_date)
df['pub_week'] = df.pub_date.map(lambda x : date.isocalendar(x)[1])
df['pub_year'] = df.pub_date.map(lambda x : date.isocalendar(x)[0])
df['pub_month'] = df.pub_date.map(lambda x : x.month)
df['pub_week_date'] = df.pub_date.map(lambda x : x.date() + datetime.timedelta(0-x.date().weekday()))# 0 -> Monday of the pub_week
df['pub_week_date_str'] = df.pub_date.map(lambda x : (x.date() + datetime.timedelta(0-x.date().weekday())).strftime("%Y-%m-%d"))# 0 -> Monday of the pub_week

df.info()

vectorizer = pkl.load( open("data/vectorizer.pkl", "rb") )
vectors = pkl.load( open("data/vectors.pkl", "rb"))
nmf = pkl.load( open("data/nmf_20.pkl", "rb"))

```

```python
W = nmf.components_#[topicN]
A = vectors.dot(W.T)

df['topic'] = list(np.argmax(A, axis=1))
df['weight'] = list(np.max(A, axis=1))

# now sort topics w.r.t number of articles per topic
# this is just renaming the topic
dg = df[['topic','headline']].groupby('topic')
x = sorted(dg.groups.keys())
y = [len(dg.groups[i]) for i in x]
m = list(np.argsort(y)[::-1])
d = { j : x[i] for i, j in enumerate(m)}

df['topic_sorted'] = df['topic'].map(lambda x : d[x])

df[['topic','topic_sorted','weight','pub_date','pub_week','pub_week_date','headline']].head()
```

```python
#word cloud

def featurize(article):
    # tokenize into words
    tokens = [word.lower() for sent in sent_tokenize(article) \
              for word in word_tokenize(sent)]

    # remove stopwords
    stop = stopwords.words('english')
    # some extra stop words not present in stopwords
    stop += ['said', 'would', 'also', 'U', 'mr', 're', 'may', 'one', 'two', 'buy', 'much', \
             'take', 'might', 'say', 'new', 'year', 'many']

    tokens = [token for token in tokens if token not in stop]

    # remove words less than three letters
    tokens = [word for word in tokens if len(word) >= 2]

    return " ".join(tokens)


n = 50   # number of articles per topic

def DrawWordCloud(df, topic, n_articles=50, save=False):
    ftopic = df[df['topic_sorted']==topic].sort('weight',ascending=False).head(n)
    text = list(ftopic['content'].values)
    text = " ".join(text)
    text = featurize(text)

    wordcloud = WordCloud().generate(text)
    plt.figure(figsize=(6,3))
    plt.imshow(wordcloud)
    plt.axis("off")
    #plt.show()
    if save:
        fileName = "topic_"+str(topic)+"temp.png"
        plt.savefig(fileName);


# draw word cloud for each topic and save png files
for i in range(20):
    DrawWordCloud(df, i, 50, True)


# trim output wordcloud images using imagemagick convert function
for i in range(20):
    d = "/home/iyounus/zipfian_project/TechNewsTrends/"
    fileName = d + "topic_" + str(i) + "temp.png"
    os.system("convert " + fileName + " -trim " + d + "topic_" + str(i) + ".png")
    os.system("rm " + fileName)
```

```python
# Create JDOS file for d3 objects

col1 = ["headline"+str(i+1) for i in range(10)]
col2 = ['url'+str(i+1) for i in range(10)]

j = []
for topic in df.topic.order().unique().tolist():
    for pub_week_date in df.pub_week_date.order().unique().tolist():
        pub_week_date_str = pub_week_date.strftime("%Y-%m-%d")
        cond = "topic == " + str(topic) + " & pub_week_date_str == '" + pub_week_date_str + "'"
        #cond = "topic == " + str(topic) + " & pub_week == " + str(pub_week)
        #print cond
        dg = df.query(cond).sort(['weight'], ascending=[0])[['topic','pub_week','pub_week_date','headline','web_url']]
        headlines = [unicodedata.normalize('NFKD', h).encode('ascii','ignore') for h in dg['headline'].values.tolist()]
        urls = dg['web_url'].values.tolist()

        row = {}
        row['N_articles'] = len(urls)
        row['pub_week_date'] = pub_week_date_str
        row['topic'] = topic
        row['pub_week'] = date.isocalendar(pub_week_date)[1]
        for i in range(len(col1)):
            if i < len(urls):
                row[col2[i]] = urls[i]
                row[col1[i]] = headlines[i]
            else:
                row[col2[i]] = "abcd"
                row[col1[i]] = "abcd"

        j.append(row)

        
with open('data_20.txt', 'w') as outfile:
  json.dump(j, outfile)
```
