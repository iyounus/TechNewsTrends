```python
import numpy as np
import pandas as pd
import pickle as pkl

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF

from nltk import WordNetLemmatizer, sent_tokenize, word_tokenize
from nltk.corpus import stopwords

from topic_ana import featurize

from datetime import date

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pkl.load( open("data/nyt_data.pkl"))
df.pub_date = pd.to_datetime(df.pub_date)
df['pub_week'] = df.pub_date.map(lambda x : date.isocalendar(x)[1])
df['pub_year'] = df.pub_date.map(lambda x : date.isocalendar(x)[0])
df['pub_month'] = df.pub_date.map(lambda x : x.month)


vectorizer = pkl.load( open("data/vectorizer.pkl", "rb") )
vectors = pkl.load( open("data/vectors.pkl", "rb"))
nmf = pkl.load( open("data/nmf_10.pkl", "rb"))
```

```python
W = nmf.components_#[topicN]
A = vectors.dot(W.T)

df['topic'] = list(np.argmax(A, axis=1))
df['weight'] = list(np.max(A, axis=1))

df[['topic','weight','pub_date','pub_week','headline']].head()
```

```python
n_top_words = 100
feature_names = vectorizer.get_feature_names()
d = {}

for j, topic in enumerate(nmf.components_):
    title = "Topic%d:" % j
    d[title] = [feature_names[i] for i in topic.argsort()[::-1]]

res = pd.DataFrame(d)
res.head(20)
```

```python
#word cloud

def featurize(article):
    # tokenize into words
    tokens = [word.lower() for sent in sent_tokenize(article) \
              for word in word_tokenize(sent)]

    # remove stopwords
    stop = stopwords.words('english')
    # some extra stop words not present in stopwords
    stop += ['said', 'would', 's', 'also', 'U', 'mr', 're', 'may']

    tokens = [token for token in tokens if token not in stop]

    # remove words less than three letters
    tokens = [word for word in tokens if len(word) >= 2]

    return " ".join(tokens)


n = 50   # number of articles per topic
from wordcloud import WordCloud
ftopic = df[df['topic']==topic].sort('weight',ascending=False).head(n)
text = list(ftopic['content'].values)
#print text
text = " ".join(text)
text = featurize(text)

wordcloud = WordCloud().generate(text)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
